```python
data = [80,20,30,60,50]
data.sort()
n=len(data)
if(n%2==0):
  median = (data[n//2-1] + data[n//2])/2
else:
  median = data[n//2]
print("Median:",median)
```

```python
data = [10,20,20,40,50]
frequency = {}
for num in data:
  if num in frequency:
    frequency[num]+=1
  else:
    frequency[num] = 1
mode = max(frequency,key=frequency.get)
print("Mode:",mode)
```
