```python
import pandas as pd
d={'one':pd.Series([1,2,3],index=['a','b','c']),
   'two':pd.Series([1,2,3,4],index=['a','b','c','d'])
   }
df=pd.DataFrame(d)
print(df)
```

```python
#Adding a new column
print("Adding a new column by passing a series:")
df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)
```

```python
#Add existing columns with each other
print("Adding a column using AN existing columns in DataFrame:")
df['four']=df['one']+df['three']
print(df)
```

```python
#Delete using DEL function
print("Deleting the first column using DEL function:")
del df['one']
print(df)
```

```python
#Delete using POP function
print("Deleting another column using POP function:")
df.pop('two')
print(df)
```

```python
#Slicing
print(df[2:4])
```

```python
print(df[1:3])
```

```python
import pandas as pd
#Addition of rows
df2=pd.DataFrame([[5,6],[7,8]],columns=['a','b'])
df=pd.concat([df, df2], ignore_index=True)
print(df)
```

```python
#Deletion of rows
df=df.drop(0)
print(df)
```

```python
#Reindexing
df1=pd.DataFrame({"A":[1,5,3,4,2],
                  "B":[3,2,4,3,4]},
                 index=["A1","A2","A3","A4","A5"])

df2=pd.DataFrame({"A":[1,5,3,4,2],
                  "B":[3,2,4,3,4]},
                 index=["A1","A2","A3","A4","A5"])

df1.reindex_like(df2)
```
